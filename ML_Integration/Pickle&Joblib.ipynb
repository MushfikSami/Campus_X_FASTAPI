```python
import pickle
import joblib
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
df=pd.read_csv('/content/House Price Prediction Dataset.csv').drop(['Location','Condition','Garage'],axis=1)
df.head()
```

```python
x=df.drop('Price',axis=1)
y=df['Price']
```

```python
model=LinearRegression()
model.fit(x,y)
```

```python
y_pred=model.predict(x)
```

```python
with open('model.pkl','wb') as f:
  pickle.dump(model,f)
```

```python
with open('model.pkl','rb') as f:
  pickle_model=pickle.load(f)
```

```python
pickle_model
```

```python
y_pred_pickle=pickle_model.predict(x)
```

```python
r2_score_pick=r2_score(y,y_pred_pickle)
r2_score_pick
```

```python
joblib.dump(model,'model_joblib')
```

```python
joblib_model=joblib.load('model_joblib')
joblib_model
```

```python
y_pred_job=joblib_model.predict(x)
```

```python
r2_score_job=r2_score(y,y_pred_job)
r2_score_job
```
